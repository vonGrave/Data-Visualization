# Tips EDA


Data Features:

**total_bill**: a numeric vector, the bill amount (dollars)


**tip**: a numeric vector, the tip amount (dollars)


**sex**: a factor with levels Female Male, gender of the payer of the bill


**smoker**: a factor with levels No Yes, whether the party included smokers


**day**: a factor with levels Friday Saturday Sunday Thursday, day of the week


**time**: a factor with levels Day Night, rough time of day


**size**: a numeric vector, number of people in a table/Party

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
tips = pd.read_csv('Tips.csv')
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.info()
```

### Understand the distribution of 'tip' and 'total_bill' using different plots

```python
sns.distplot(tips['total_bill'],color='green')
plt.show()
sns.boxplot(y='total_bill',data=tips)
plt.show()
```

```python
sns.distplot(tips['tip'],color='blue')
plt.show()
sns.boxplot(tips['tip'],orient='v')
plt.show()
```

Both 'tip' and 'total_bill' are skewed towards right but 50 % people paying less than 20 dollars for overall bill and 50 % people paying less than 3 dollars as tip

### Relationship between 'total_bill' and 'tip'

Implement jointplot, hex jointplot, pairplot and lmplot

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind="reg", color='red')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind="hex", color='green')
plt.show()
```

```python
sns.pairplot(tips[['total_bill','tip']])
plt.show()
```

```python
sns.lmplot(x = "total_bill",y = "tip", data = tips)
plt.show()
```

A linear realtionship can be observed as the bill amount increases so does the tip

### Make plots to check distribution of tip with different gender.

```python
sns.stripplot(data=tips, x='sex', y='tip', hue='sex')
plt.show()
```

```python
sns.boxplot(data=tips, x='sex', y='tip')
plt.show()
```

Median value for tip is almost same for both, but male have a lot of outliers giving higher tip

### Draw relationship between 'total_bill', 'tip' and 'sex'

using lmplot

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='sex')
plt.show()
```

Outliers with higher billing amount and tip are males

### Which day brings in higher bill amount?

using stripplot, swarmplot and boxplot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False, hue='day')
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips, hue='day')
plt.show()
```

```python
sns.boxplot(y='total_bill',x='day',data=tips)
plt.show()
```

'Sat' and 'Sun' seems to be good days as total bill and number of customers are high on those days

### Find out count of total males and females visiting each day.

using countplot

```python
sns.countplot(x=tips['day'], hue=tips['sex'])
plt.show()
```

### At what time of which day is the total bill highest with respect to tips (consider for male and female separately)

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='day',row='sex',col='time')
plt.show()
```

Both male and female spend more at Dinners on Sat and Sun

### Plot graphs for smoker and non-smoker (try drawing relationship with 'tip').

using stripplot and boxplot

```python
sns.stripplot(data=tips, x='smoker', y='tip', hue='smoker')
plt.show()
```

```python
sns.boxplot(y='tip',x='smoker',data=tips)
plt.show()
```

No relation between smoking and tips, median is almost same

### Draw relationship between 'total_bill', 'tip' and 'smokers'

```python
sns.countplot(x=tips['smoker'])
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='sex',col='smoker',palette='rocket')
plt.show()
```

Highest tip was given by a smokers

Non smoker show a more prominent linear relationship between total_bill and tip compared to smokers
